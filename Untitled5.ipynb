### 1.1. Скачать следующие данные: kc-house-data и laptop_price

```python
df = pd.read_csv('kc-house-data.csv')
```

```python
df['delta_renovated'] = (df['yr_renovated'] - df['yr_built']).apply(lambda year: year if year > 0 else 0)
df['Year'] = df['date'].apply(lambda date: datetime.strptime(date, '%Y%m%dT%H%M%S').year)
df['Month'] = df['date'].apply(lambda date: datetime.strptime(date, '%Y%m%dT%H%M%S').month)
df.drop(columns=['date', 'zipcode', 'lat', 'long'], inplace=True)
```

###1.2. Изучите стоимости недвижимости

```python
plt.figure(figsize=(16, 9))
plt.hist(df['price'], bins=100, edgecolor='black')
plt.title("House's price")
plt.xlabel('Price')
plt.ylabel('Quantity')
plt.xticks(range(50000, 8000000, 250000));
```

Больше всего домов стоимостью около 500000.
Мало домов стоимостью выше 1000000.

### 1.3. Изучите распределение квадратуры жилой

```python
plt.figure(figsize=(16, 9))
plt.hist(df['sqft_living'], bins = 100, edgecolor='black')
plt.title("Square living")
plt.xlabel('Square')
plt.ylabel('Quantity')
plt.xticks(range(0, 10000, 500));
```

Большинство домов площадью от 700 до 4700 кв футов. больше всего домов имеют площадь 1500 кв футов.

### 1.4. Изучите распределение года постройки

```python
plt.figure(figsize=(16, 9))
plt.hist(df['yr_built'], bins = 100, edgecolor='black')
plt.title("Square living")
plt.xlabel('Year of built')
plt.ylabel('Quantity')
plt.xticks(range(df['yr_built'].min(), df['yr_built'].max(), 5));
```

Много домов построенно в 1947, 1953, 1962, 1976, 1984, 1999, 2007 и 2014 годах, возможно вызвано ставками по ипотеке

### 2.3. Изучите распределение состояния домов
